# 4. Plots:

```r
#1. Avg salary plot for 2013 to 2017
salary_2013<-mean(dataset[dataset$Year=="2013",c("Total.Salary")])
salary_2014<-mean(dataset[dataset$Year=="2014",c("Total.Salary")])
salary_2015<-mean(dataset[dataset$Year=="2015",c("Total.Salary")])
salary_2016<-mean(dataset[dataset$Year=="2016",c("Total.Salary")])
salary_2017<-mean(dataset[dataset$Year=="2017",c("Total.Salary")])
avg_salary<-c(salary_2013,salary_2014,salary_2015,salary_2016,salary_2017) 
year<-c(2013,2014,2015,2016,2017)
barplot(avg_salary,names.arg=year,xlab="Year",ylab="Avg. salary",col="green",main="Avg. salary  2013 to 2017")


```

The above bar plot shows the change of average salary over the time period 2013 to 2017. In 2014, the average salary dropped by -1.19% that salary dropped continue in the year 2015 too. The situation has been changed in 2016. In 2016, it has increased to 5% by comparison to 2015. Also, the average salary has risen in 2017 too. 

# 4.2 Plots:

```r
# Total Compensation over 5 years
Total.Compensation_2013<-mean(dataset[dataset$Year=="2013",c("Total.Compensation")])
Total.Compensation_2014<-mean(dataset[dataset$Year=="2014",c("Total.Compensation")])
Total.Compensation_2015<-mean(dataset[dataset$Year=="2015",c("Total.Compensation")])
Total.Compensation_2016<-mean(dataset[dataset$Year=="2016",c("Total.Compensation")])
Total.Compensation_2017<-mean(dataset[dataset$Year=="2017",c("Total.Compensation")])
Total.Compensation<-c(Total.Compensation_2013,Total.Compensation_2014,Total.Compensation_2015,Total.Compensation_2016,Total.Compensation_2017) 
year<-c(2013,2014,2015,2016,2017)

plot(year,Total.Compensation, type = "l", lty = 1,xlab = "Year", ylab = "Avg. Total Compensation",
   main = "The change of Avg. Total Compensation over 5 years")

```

The above line chart shows the variation of average total compensation data between 2013 to 2017.The average total compensation has faced a slight increment between 2013 to 2015. But it has shown a sharp rise in the average total compensation in 2016. Also, it has followed the trend in 2017 too. The overall trend was helpful for the employee. They have received more benefits from the employer.  

# 4.3 Plots:

```r
overtime<-dataset[dataset$Overtime=="0",c("Overtime")]
No_overtime_opportunities_employees= length(overtime)/length(dataset$Overtime)
Overtime_opportunities_employees=1-No_overtime_opportunities_employees
slices <- c(No_overtime_opportunities_employees,Overtime_opportunities_employees) 
lbls <- c("No overtime opportunities employees,", "Overtime opportunities employees,")
pct <- round(slices/sum(slices)*100)
lbls <- paste(lbls, pct) # add percents to labels 
lbls <- paste(lbls,"%",sep="") # ad % to labels 
pie(slices,labels = lbls, col=rainbow(length(lbls)),
   main="Employees Overtime Opportunities")
```

The above pie chart represents the percentage of the employee who has overtime opportunities. The overtime opportunity has brought the extra possibility to earn more wage. Those extra earnings have helped them to cope with the city's high living cost.
